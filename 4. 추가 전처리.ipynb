```python
import pandas as pd
import numpy as np
data_dir = './data/'

df_1 = pd.read_feather(data_dir + 'processed_data_1901_2004_ver1.ftr')
df_2 = pd.read_feather(data_dir + 'processed_data_1901_2004_ver2.ftr')

info_features = df_1.loc[:, :'연평균소득추정'].columns.tolist()
features = df_1.loc[:,'가정용품':'주거'].columns.tolist()

df_1
```

## 100% 제거
- 원본: 33만 7034
- 100% 제거: 31만 2545
- 95% 제거: 31만 138
- 90% 제거: 30만 6593
- 50% 제거: 22만 9322

```python
def remove_outliers(df, percent):

  index_to_remove = []

  for f in features:
    index_to_remove.extend(df[df[f] >= percent].index.values.tolist())

  return df.drop(index_to_remove)
```

```python
df_1 = remove_outliers(df_1, 1)
df_2 = remove_outliers(df_2, 1)

df_1
```

### 중복된 '가맹점소재지2' 체크
- 청원군: 2014년 청주시와 통합됨

```python
def check_duplicate(df, region2_list):
    for region in region2_list:
        print(region, ': ', df[df['가맹점소재지2'] == region]['가맹점소재지1'].unique())

region2_list = ['양산시', '김해시', '창원시',
               '경산시', '영천시', '군위군', '청도군', '고령군', '성주군', '칠곡군', '창녕군',
               '나주시', '담양군', '화순군', '함평군', '장성군',
               '공주시', '논산시', '계룡시', '금산군', '청주시', '보은군', '청원군', '옥천군']
check_duplicate(df_1, region2_list)
```

```python
import warnings
warnings.filterwarnings(action='ignore')

def label_regions(df):

    all_data = []

    # 1. 수도권

    region1 = ['서울특별시', '인천광역시', '경기도']

    df_filtered = df[df['가맹점소재지1'].isin(region1)]
    df_filtered['지역'] = '수도권'
    all_data.append(df_filtered)
    count = df_filtered.shape[0]

    # 2. 부산울산권

    region1 = ['부산광역시', '울산광역시']
    region2 = ['양산시', '김해시', '창원시']

    df_filtered = df[df['가맹점소재지1'].isin(region1) | df['가맹점소재지2'].isin(region2)]
    df_filtered['지역'] = '부산울산권'
    all_data.append(df_filtered)
    count += df_filtered.shape[0]

    # 3. 대구권

    region1 = ['대구광역시']
    region2 = ['경산시', '영천시', '군위군', '청도군', '고령군', '성주군', '칠곡군', '창녕군']

    df_filtered = df[df['가맹점소재지1'].isin(region1) | df['가맹점소재지2'].isin(region2)]
    df_filtered['지역'] = '대구권'
    all_data.append(df_filtered)
    count += df_filtered.shape[0]

    # 4. 광주권

    region1 = ['광주광역시']
    region2 = ['나주시', '담양군', '화순군', '함평군', '장성군']

    df_filtered = df[df['가맹점소재지1'].isin(region1) | df['가맹점소재지2'].isin(region2)]
    df_filtered['지역'] = '광주권'
    all_data.append(df_filtered)
    count += df_filtered.shape[0]

    # 5. 대전권

    region1 = ['대전광역시', '세종특별자치시']
    region2 = ['공주시', '논산시', '계룡시', '금산군', '청주시', '보은군', '청원군', '옥천군']

    df_filtered = df[df['가맹점소재지1'].isin(region1) | df['가맹점소재지2'].isin(region2)]
    df_filtered['지역'] = '대전권'
    all_data.append(df_filtered)
    count += df_filtered.shape[0]
    
    ###
    temp = pd.concat(all_data, axis=0, ignore_index=False)
    regions_so_far = temp['가맹점소재지2'].unique()
    ###
    
    # 6. 나머지 

    df_filtered = df[~df['가맹점소재지2'].isin(regions_so_far)]
    df_filtered['지역'] = '기타'
    all_data.append(df_filtered)
    count += df_filtered.shape[0]
    
    # 최종

    print('total rows: ', count)
    return pd.concat(all_data, axis=0, ignore_index=False) 


df_1_labeled = label_regions(df_1)
df_2_labeled = label_regions(df_2)
```

```python
sorted_features = info_features+['지역']+features
```

```python
df_1_labeled = df_1_labeled[sorted_features]
df_2_labeled = df_2_labeled[sorted_features]

df_1_labeled.reset_index(inplace=True)
df_2_labeled.reset_index(inplace=True)

df_1_labeled.drop(columns=['index']).to_feather(
    data_dir + 'clustering_data_1901_2004_ver1.ftr')
df_2_labeled.drop(columns=['index']).to_feather(
    data_dir + 'clustering_data_1901_2004_ver2.ftr')
```
